Определите функцию разметки на основе правил, которая присваивает метки на основе определенных условий или закономерностей в данных.

Примените функцию разметки для автоматической разметки подмножества данных.

```python
import pandas as pd

data = pd.read_csv("googleplaystore.csv")

data.head()
```

```python
def rule_based_labeling(row):
    if row["Rating"] >= 4:
        return "Good"
    elif row["Rating"] > 0 and row["Rating"] < 4:
        return "Bad"
    else:
        return "No rating"
```

```python
data["Rating_Cat"] = data.apply(rule_based_labeling, axis=1)
data.head()
```
